#### Question 4: 
Split the data into train and test data, What parameters do you use for splitting the data? Why? Do you apply stratified sampling? Why or Why not?

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
strat_train_set, strat_test_set = train_test_split(df, test_size=0.2, random_state=42,
                                         stratify = df["room_type"])
```

```python
def room_type_proportions(data):
    return data["room_type"].value_counts() / len(data)

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

proportion_compare = pd.DataFrame({
    "Overall": room_type_proportions(df),
    "Stratified": room_type_proportions(strat_test_set), # train_test_split
    "Random": room_type_proportions(test_set),
}).sort_index()
proportion_compare["Rand. %error"] = 100 * proportion_compare["Random"] / proportion_compare["Overall"] - 100
proportion_compare["Strat. %error"] = 100 * proportion_compare["Stratified"] / proportion_compare["Overall"] - 100
proportion_compare
```

<font color='blue'><b>Parameters:</b> test_size is 0.2, random_state(42) and stratify = df["room_type"] are being used for splitting the data.<br></font>
<font color='blue'><b>stratified sampling</b> While inspecting the categorical features, we have observed that the "room_type" has three unique values in which the "Shared room" has less incidence relative to other values so in this case the Startified Sampling work better than random sample.Unlike The Simple random, In Stratified random sample we split the data in proportion to the group's size instead of whole data.</font>
    
<font color='blue'>For exapmple here, room_type has three different sub groups with different proportions in the size. Stratified random sample splits the data according to their individual sizes.</font>
    
<font color='blue'>Entire home/apt = 25409<br>Private room    = 22326<br>Shared room     = 1160<br></font>

<font color='blue'>Also From the above table, stratified and random samples with their Strat. %error and Rand. %error shows that the stratified sample has less error percentage. </font>

#### Question 6: 
Use Pipeline, ColumnTransformer and OneHotEncoder to fit and evaluate Linear Regression and Decision Tree models with default parameters to predict the "price". Discuss your results and findings.


<font color='blue'>We are creating the data frames (x_train_num and x_train_cat) based on the type of the data like numeric and categorical. </font>

```python
#Selecting numerical dataframe in train set
x_train_num=x_train.select_dtypes(include=np.number)

#Selecting categorical dataframe in train set
x_train_cat=x_train.select_dtypes(exclude=['number'])
```

Usually OneHotEncoder returns a sparse array, we are using toarray() method to conevrt it into dense array. we can also acheive this by setting the sparse parameter as False in OneHotEncoder(sparse=False).</font>  

```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
x_train_cat_hot = cat_encoder.fit_transform(x_train_cat)
x_train_cat_hot.toarray()
```

## Transformation of Pipelines

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

#num_pipeline=Pipeline([('imputer',SimpleImputer(strategy='median')),])

num_pipeline=Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value=0)),])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#numerical and categorical attributes
num_attribs = list(x_train_num.columns)
cat_attribs = list(x_train_cat.columns)

#full pipeline of numeric and categorical attributes
full_pipeline = ColumnTransformer([("num",num_pipeline,num_attribs),("cat",OneHotEncoder(),cat_attribs)])

#Using full_pipeline perform fit and transform on training data
hotel_pipeline = full_pipeline.fit_transform(x_train)
```

```python
x_train_pipeline.shape
```

```python
full_pipeline
```

## LinearRegression

```python
#importing libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score, mean_absolute_error

#linear regressor Definition
linear_reg_mod=LinearRegression()

#model fitting on x_train and y_train data
linear_model = linear_reg_mod.fit(hotel_pipeline,y_train)

#prediction on x_train
linear_mod_predictions = linear_model.predict(hotel_pipeline)

#MSE and RMSE values
linear_model_mse=mean_squared_error(y_train,linear_mod_predictions)
linear_model_rmse=np.sqrt(linear_model_mse)

#MAE values
linear_model_mae=mean_absolute_error(y_train,linear_mod_predictions)

#r2 values
linear_model_r2=r2_score(y_train,linear_mod_predictions)

print("Mean squared error(MSE): %.3f" % linear_model_mse)
print("Root mean squared error(RMSE): %.3f" % linear_model_rmse)
print("Mean Absolute error(MAE): %.3f" % linear_model_mae)
print("R2 score: %.3f" % linear_model_r2)
```

<font color='blue'>For a regression problem performence of the model can be measured using Root mean squared error(RMSE), which indicates how much error the model makes while predicting. In few cases like when there are outliers, we can consider Mean Absolute Error(MAE) which corresponds to manhattan distance. where as RMSE corresponds to Euclidean distance. </font>

<font color='blue'>We can obeserve the RMSE value is 0.472 and MAE value is 0.338 which is very high with r2 score 0.539. As expected linear model struggles to fit non-linear data.</font>

<font color='blue'>As result the model performence is not quite good.</font>

## DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

#Decision tree regressor
dec_tree_reg=DecisionTreeRegressor()

#model fitting on x_train and y_train data
dec_tree_reg.fit(hotel_pipeline,y_train)

#predictions on x_train
dec_tree_predictions = dec_tree_reg.predict(hotel_pipeline)

#MSE and RMSE values
dec_tree_mse=mean_squared_error(y_train,dec_tree_predictions)
dec_tree_rmse=np.sqrt(dec_tree_mse)

#R2 Score
dec_tree_r2=r2_score(y_train,dec_tree_predictions)

print("Mean squared error(MSE): %.3f" % dec_tree_mse)
print("Root mean squared error(RMSE): %.3f" % dec_tree_rmse)
print("R2 score: %.3f" % dec_tree_r2)
```

<font color='blue'>As per the result it is observed that the model is overfittling. So we would like to use parameter tuning in the model to minimize the overfitting. </font>

## Decision tree regressor with parameter tuning

```python
#Decision tree regressor
dec_tree_reg=DecisionTreeRegressor(max_depth=20, splitter = "best")

#model fitting on x_train and y_train data
dec_tree_reg.fit(hotel_pipeline,y_train)

#predictions on x_train
dec_tree_predictions = dec_tree_reg.predict(hotel_pipeline)

#MSE and RMSE values
dec_tree_mse=mean_squared_error(y_train,dec_tree_predictions)
dec_tree_rmse=np.sqrt(dec_tree_mse)

#R2 Score
dec_tree_r2=r2_score(y_train,dec_tree_predictions)

print("Mean squared error(MSE): %.3f" % dec_tree_mse)
print("Root mean squared error(RMSE): %.3f" % dec_tree_rmse)
print("R2 score: %.3f" % dec_tree_r2)
```

<font color='blue'>To over come the overfitting problem, we tried with parameters max_depth and splitter and observed prediction varies based on different combinations. We think this model performance is better than previous overfitting model.</font>
